## Linearly separable and not linearly datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons

# Linearly separable dataset (using make_blobs)
def generate_linearly_separable():
    # Generate two clusters (blobs) that are linearly separable
    X, y = make_blobs(n_samples=200, centers=2, n_features=2, random_state=42, cluster_std=1.5)
    return X, y

# Non-linearly separable dataset (using make_moons)
def generate_non_linearly_separable():
    # Generate a dataset where two classes form interlocking half circles (moons)
    X, y = make_moons(n_samples=200, noise=0.1, random_state=42)
    return X, y

# Plot the datasets
def plot_datasets(X_lin, y_lin, X_nonlin, y_nonlin):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Plot the linearly separable dataset
    axes[0].scatter(X_lin[y_lin == 0][:, 0], X_lin[y_lin == 0][:, 1], color='red', label='Class 0')
    axes[0].scatter(X_lin[y_lin == 1][:, 0], X_lin[y_lin == 1][:, 1], color='blue', label='Class 1')
    axes[0].set_title('Linearly Separable Dataset')
    axes[0].legend()
    axes[0].grid(True)

    # Plot the non-linearly separable dataset
    axes[1].scatter(X_nonlin[y_nonlin == 0][:, 0], X_nonlin[y_nonlin == 0][:, 1], color='red', label='Class 0')
    axes[1].scatter(X_nonlin[y_nonlin == 1][:, 0], X_nonlin[y_nonlin == 1][:, 1], color='blue', label='Class 1')
    axes[1].set_title('Non-Linearly Separable Dataset')
    axes[1].legend()
    axes[1].grid(True)

    plt.show()

# Generate the datasets
X_lin, y_lin = generate_linearly_separable()
X_nonlin, y_nonlin = generate_non_linearly_separable()

# Plot the datasets
plot_datasets(X_lin, y_lin, X_nonlin, y_nonlin)
```

## Common activation functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def perceptron(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input data
x = np.linspace(-10, 10, 100)

# Calculate activations
y_perceptron = perceptron(x)
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(12, 10))

# Plot Perceptron
plt.subplot(2, 2, 1)
plt.plot(x, y_perceptron, label='Perceptron Activation')
plt.title('Perceptron Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

# Plot Sigmoid
plt.subplot(2, 2, 2)
plt.plot(x, y_sigmoid, label='Sigmoid Activation', color='orange')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

# Plot ReLU
plt.subplot(2, 2, 3)
plt.plot(x, y_relu, label='ReLU Activation', color='green')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.ylim(-1, 10)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

# Plot Tanh
plt.subplot(2, 2, 4)
plt.plot(x, y_tanh, label='Tanh Activation', color='purple')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.ylim(-1.1, 1.1)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

plt.tight_layout()
plt.show()
```
